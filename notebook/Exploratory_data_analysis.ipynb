```python
# Define numerical and categorical columns

numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O'] # 'O' denotes object type of the data in the column which means text data
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# Print colunms
print('We have {} numerical features: {}'.format(len(numerical_features), numerical_features))
print('We have {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

***Total score and Average Score***

```python
# Total score
df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']

# Average score
# np.round() roundoff the decimal to 2 places
df['average'] = np.round(df['total_score']/3,2)
```

```python
df.head()
```

```python
# Getting students with perfect score of 100 score in each of the subjects

reading_full = df[df['reading_score']== 100]['average'].count()
writing_full = df[df['writing_score']== 100]['average'].count()
math_full = df[df['math_score']== 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
# Getting students with scores below 20 in each of the subjects

reading_less_than_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_than_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_than_20 = df[df['math_score'] <= 20]['average'].count()

print(f"Number of students with less than 20 marks Maths: {math_less_than_20}")
print(f"Number of students with less than 20 marks Writing: {writing_less_than_20}")
print(f"Number of students with less than 20 marks Reading: {reading_less_than_20}")
```

***Insights***

- Students performed worst in Maths
- Students performed best in reading

## Exploring Data (Visualization)

### Visualization of Average score distribution and Total Score Distribution

#### By Gender

```python
# Histogram and KDE (Kernel Distribution Function) for Average Score

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,7), tight_layout=True)

# First plot - Overall distribution
sns.histplot(data=df, x='average', bins=30, kde=True, color='g', ax=ax1)
ax1.set_title('Overall Score Distribution')
ax1.set_xlabel('Average Score')

# Second Plot - Distribution by gender
sns.histplot(data=df, x='average', kde=True, hue='gender', ax=ax2)
ax2.set_title('Score Distribution by Gender')
ax2.set_xlabel('Average Score')

plt.show()
```

**fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), tight_layout=True)** 

+ Creates a figure with 1 row and 2 columns of subplots

+ figsize=(15, 7): Makes the figure 15 inches wide and 7 inches tall

+ tight_layout=True: Automatically adjusts spacing between plots to prevent overlap

+ Returns:

    - fig: The overall figure container

    - (ax1, ax2): Tuple of Axes objects for the two subplots

**sns.histplot(data=df, x='average', bins=30, kde=True, color='g', ax=ax1)  
ax1.set_title('Overall Score Distribution')  
ax1.set_xlabel('Average Score')**
+ sns.histplot() creates a histogram with density curve:

+ data=df: Uses DataFrame df as data source

+ x='average': Plots the 'average' column

+ bins=30: Uses 30 bins for the histogram

+ kde=True: Adds Kernel Density Estimation smooth curve

+ color='g': Makes the histogram green

+ ax=ax1: Draws on the first subplot

+ set_title() and set_xlabel() add descriptive labels  

+ hue='gender': Splits the data by gender (creates overlapping histograms)

```python
# Histogram and KDE for Total score

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,7), tight_layout=True)

# First plot - Overall distribution
sns.histplot(data=df, x='total_score', bins=30, kde=True, color='r', ax=ax1)
ax1.set_title('Overall Score Distribution')
ax1.set_xlabel('Total Score')

# Second plot - Distribution by gender
sns.histplot(data=df, x='total_score', kde=True, hue='gender', ax=ax2)
ax2.set_title('Score Distribution by Gender')
ax2.set_xlabel('Total Score')

plt.show()
```

***Insights***

- Female students tend to perform better than male students

#### By Lunch

```python
# Histogram and KDE (Kernel Distribution Function) for Average Score

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(25,6), tight_layout=True)

# First plot - Overall distribution
sns.histplot(data=df, x='average', kde=True, hue='lunch', ax=ax1)
ax1.set_title('Overall Score Distribution')
ax1.set_xlabel('Average Score')

# Second Plot - Distribution by gender
sns.histplot(data=df[df['gender']=='female'], x='average', kde=True, hue='lunch', ax=ax2)
ax2.set_title('Score Distribution by Gender - Female')
ax2.set_xlabel('Average Score')

# Third Plot - Distribution by gender
sns.histplot(data=df[df['gender']=='male'], x='average', kde=True, hue='lunch', ax=ax3)
ax3.set_title('Score Distribution by Gender - Male')
ax3.set_xlabel('Average Score')


plt.show()
```

```python
# Histogram and KDE (Kernel Distribution Function) for Average Score

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(25,6), tight_layout=True)

# First plot - Overall distribution
sns.histplot(data=df, x='average', kde=True, hue='parental_level_of_education', ax=ax1)
ax1.set_title('Overall Score Distribution')
ax1.set_xlabel('Average Score')

# Second Plot - Distribution by gender
sns.histplot(data=df[df['gender']=='female'], x='average', kde=True, hue='parental_level_of_education', ax=ax2)
ax2.set_title('Score Distribution by Gender - Female')
ax2.set_xlabel('Average Score')

# Third Plot - Distribution by gender
sns.histplot(data=df[df['gender']=='male'], x='average', kde=True, hue='parental_level_of_education', ax=ax3)
ax3.set_title('Score Distribution by Gender - Male')
ax3.set_xlabel('Average Score')


plt.show()
```

#### Maximum Score of Students in Three Subjects

```python
plt.figure(figsize=(18,8))

# Math Scores
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
plt.ylabel('Scores')
sns.violinplot(y='math_score', x='gender', data=df, color='red', linewidth=3, inner='quartile')

# Reading Scores
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading_score', x= 'gender',data=df, color='green', linewidth=3, inner='quartile')

# Writing Scores
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing_score', x='gender', data=df, color='blue', linewidth=3, inner='quartile')

# Combined Boxplot (4th subplot)
plt.subplot(1, 4, 4)
plt.title('ALL SCORES COMPARISON')
sns.boxplot(data=df[['math_score', 'reading_score', 'writing_score']], palette='pastel')

plt.suptitle('Student Score Distributions', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

***Insights***

```python
# Sets large figure size 30 inch wide and 12 inch in length
plt.rcParams['figure.figsize'] = (30, 12)

# labels removed to declutter the chart

# Gender
plt.subplot(1, 5, 1)
size = df['gender'].value_counts()
labels = 'Female', 'Male'
color = ['red','green']

plt.pie(size, colors = color,autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution', fontsize = 20)
plt.axis('off')
plt.legend(size.index, bbox_to_anchor=(0.5, -0.1), loc='upper center')

# Race_Ethnicity
plt.subplot(1, 5, 2)
size = df['race_ethnicity'].value_counts()
labels = 'Group C', 'Group D','Group B','Group E','Group A'
color = ['red', 'green', 'blue', 'cyan','orange']

plt.pie(size, colors = color,autopct = '%1.1f%%', startangle=90)
plt.title('Race/Ethnicity Distribution', fontsize = 20)
plt.axis('off')
plt.legend(size.index, bbox_to_anchor=(0.5, -0.1), loc='upper center')


# Lunch
plt.subplot(1, 5, 3)
size = df['lunch'].value_counts()
labels = 'Standard', 'Free'
color = ['red','green']

plt.pie(size, colors = color,autopct = '%1.1f%%', startangle=90)
plt.title('Lunch', fontsize = 20)
plt.axis('off')
plt.legend(size.index, bbox_to_anchor=(0.5, -0.1), loc='upper center')


# Test preparation course
plt.subplot(1, 5, 4)
size = df['test_preparation_course'].value_counts()
labels = 'None', 'Completed'
color = ['red','green']

plt.pie(size, colors = color,autopct = '%1.1f%%', startangle=90)
plt.title('Test Course', fontsize = 20)
plt.axis('off')
plt.legend(size.index, bbox_to_anchor=(0.5, -0.1), loc='upper center')


# Parental level of Education
plt.subplot(1, 5, 5)
size = df['parental_level_of_education'].value_counts()
labels = 'Some College', "Associate's Degree",'High School','Some High School',"Bachelor's Degree","Master's Degree"
color = ['red', 'green', 'blue', 'cyan','orange','grey']

plt.pie(size, colors = color,autopct = '%1.1f%%', startangle=90)
plt.title('Parental Education', fontsize = 20)
plt.axis('off')
plt.legend(size.index, bbox_to_anchor=(0.5, -0.1), loc='upper center')

plt.tight_layout()
plt.grid()

plt.show()
```

### Feature Wise Visual Analysis

#### Gender Column

- What is gender wise distribution of students?
- Does gender of a student have an impact on the performance?

##### Univariate Analysis (Gender distribution)

```python
# Set figure parameters
plt.rcParams['font.size'] = 14
f, ax = plt.subplots(1,2, figsize = (20,10))
f.suptitle('Gender Distribution Comparison', y = 1.02, fontsize=18)

# Custom Colours for genders
colors = {'male': '#ff4d4d', 'female': '#ff8000'}

# Count Plot
sns.countplot(x = df['gender'], data = df, palette=colors.values(), ax=ax[0], saturation=0.95)
ax[0].set_title('Count of Students by Gender', pad=20)
ax[0].set_xlabel('Gender')
ax[0].set_ylabel('Count')

# Add value labels
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=16, padding=3)

# Pie Chart
gender_counts = df['gender'].value_counts()
ax[1].pie(gender_counts, 
          labels=gender_counts.index,
          explode=[0, 0.1],
          autopct='%1.1f%%',
          shadow=True,
          colors=colors.values(),
          startangle=90)
ax[1].set_title('Percentage Distribution', pad=20)

plt.tight_layout()
plt.show()
```

***Insights***

- Gender distribution is balanced

##### Bivariate Analysis (Impact of gender on performance)

```python
gender_group =  df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(12, 8))  # Slightly wider for better label spacing

# Data Preparation
categories = ['Total Average', 'Math Average']
female_scores = [gender_group.loc['female', 'average'], gender_group.loc['female', 'math_score']]
male_scores = [gender_group.loc['male', 'average'], gender_group.loc['male', 'math_score']]

# Positioning
x_axis = np.arange(len(categories))
bar_width = 0.35  # Adjusted for better spacing

# Plotting
plt.bar(x_axis - bar_width/2, male_scores, bar_width, 
        label='Male', color='#1f77b4', edgecolor='black')
plt.bar(x_axis + bar_width/2, female_scores, bar_width, 
        label='Female', color='#ff7f0e', edgecolor='black')

# Customization
plt.xticks(x_axis, categories, fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Average Score", fontsize=14, labelpad=10)
plt.title("Gender Comparison: Total vs Math Averages", 
          fontsize=16, fontweight='bold', pad=20)

# Add value labels on top of bars
for i, (male, female) in enumerate(zip(male_scores, female_scores)):
    plt.text(i - bar_width/2, male + 0.5, f'{male:.1f}', 
             ha='center', va='bottom', fontsize=12)
    plt.text(i + bar_width/2, female + 0.5, f'{female:.1f}', 
             ha='center', va='bottom', fontsize=12)

# Legend and grid
plt.legend(fontsize=12, framealpha=1)
plt.grid(axis='y', alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

***Insights***

- On average female students perform better than male students
- Male students have higher scores in Maths
